### Yield and Growth factors

### Yield

```python
import requests
from bs4 import BeautifulSoup as bs
import re
import itertools
```

```python
url ='https://www.multpl.com/s-p-500-dividend-yield/table/by-month'
response=requests.get(url)
page=response.text
soup=bs(page,'html.parser')
```

```python
year=list(range(1999,2019))
l=[]
for y in year:
    date_list=soup.find_all(text=re.compile(str(y)))
    l.append(date_list)
    
l
```

```python
all_dates=list(itertools.chain(l[0],l[1],l[2],l[3],l[4],
                               l[5],l[6],l[7],l[8],l[9],
                               l[10],l[11],l[12],l[13],l[14],
                               l[15],l[16],l[17],l[18],l[19]))   
all_dates=[str(e) for e in all_dates]
all_dates
```

```python
yields=[]

for d in all_dates:
    yield_text=soup.find(text=d).next.next.text
    yield_text=re.match('[\d.]+',yield_text)[0]
    yields.append(yield_text)
    
yields
```

```python
k={}
k['year']=all_dates
k['yields']=yields

div_yield=pd.DataFrame(k)
div_yield.columns=['Date','Dividend yield']
div_yield
```

```python
div_yield.set_index('Date',inplace=True)
div_yield.index=pd.to_datetime(div_yield.index)
div_yield=div_yield.sort_index()['1999-03-31':'2018-11-30'].apply(lambda x: pd.to_numeric(x)).div(100)
div_yield
```

### Earnings growth

```python
link ='https://www.multpl.com/s-p-500-real-earnings-growth/table/by-quarter'
p=requests.get(link).text
earnings_soup=bs(p,'html.parser')
```

```python
year_range=list(range(1999,2019))
w=[]
for i in year_range:
    d_list=earnings_soup.find_all(text=re.compile(str(i)))
    w.append(d_list)
    
w
```

```python
all_qtr=list(itertools.chain(w[0],w[1],w[2],w[3],w[4],
                             w[5],w[6],w[7],w[8],w[9],
                             w[10],w[11],w[12],w[13],w[14],
                             w[15],w[16],w[17],w[18],w[19]))
all_qtr=[str(e) for e in all_qtr]
all_qtr
```

```python
growth=[]

for q in all_qtr:
    earnings_text=earnings_soup.find(text=q).next.next.text
    earnings_text=re.match('[-\d.]+',earnings_text)[0]
    growth.append(earnings_text)
    
growth
```

```python
v={}
v['Date']=all_qtr
v['Earnings growth']=growth

earnings_growth=pd.DataFrame(v)
earnings_growth
```

```python
earnings_growth.set_index('Date',inplace=True)
earnings_growth.index=pd.to_datetime(earnings_growth.index)
earnings_growth=earnings_growth.sort_index()
earnings_growth=earnings_growth.apply(lambda x: pd.to_numeric(x)).div(100)
earnings_growth
```

```python
earnings_growth=earnings_growth.apply(lambda x: (1+x/4).pow(1/3)-1)
earnings_growth=earnings_growth.resample('M').bfill()['1999-03-31':'2018-11-30']
earnings_growth
```

## Capital Asset Pricing Model (CAPM)

```python
df = pd.read_csv('raw_data.csv')
df
```

```python
df = df.set_index(pd.DatetimeIndex(df['Date']))
```

```python
df
```

```python
rf=web.DataReader('^IRX','yahoo',start=start,end=end)
rf=rf['Adj Close']
rf=(1+(rf/100))**(1/12)-1
rf=rf.resample('M').mean()
rf
```

```python
spx=web.DataReader('SPY','yahoo',start=start,end=end)
spx=spx['Adj Close']
spx=spx.resample('M').mean()
spx
```

```python
spx_rt=spx.pct_change()
mkt_excessrt=spx_rt-rf
mkt_excessrt.fillna(0,inplace=True)
mkt_excessrt
```

```python
df=df.resample('M').mean()
df_monthrt=df.pct_change().fillna(0)
df_monthrt
```

```python
df_excessrt= df_monthrt.apply(lambda x: x-rf)
df_excessrt
```

```python
df_excessrt=df_excessrt.apply(lambda x: np.nan_to_num(np.array(x)))
df_excessrt
```

```python
alphas={}
for i in df_excessrt.columns.unique():
    y=np.array(df_excessrt[i])
    x=np.array(mkt_excessrt).reshape(237,1)
    
    model=linear_model.LinearRegression().fit(x,y)
    
    alphas[i]=model.intercept_
    
alphas
```

```python
data={}
tickers=[]
capm_alphas=[]
for i in df_excessrt.columns.unique():
    y=np.array(df_excessrt[i])
    x=np.array(mkt_excessrt).reshape(237,1)
    
    model=linear_model.LinearRegression().fit(x,y)
    
    tickers.append(i)
    capm_alphas.append(model.intercept_)
    
data['Tickers']=tickers
data['Alphas']=capm_alphas

capm_df=pd.DataFrame(data)
capm_df
```

```python
capm_sorted=capm_df.sort_values(['Alphas'],ascending=False)
capm_sorted
```

```python
capm_top10=capm_sorted[capm_sorted.Alphas>=capm_sorted.Alphas.quantile(0.98)]
capm_top10
```

```python
capm_bottom10=capm_sorted[capm_sorted.Alphas<=capm_sorted.Alphas.quantile(0.02)]
capm_bottom10
```

```python
capm_top50=capm_sorted[capm_sorted.Alphas>=capm_sorted.Alphas.quantile(0.9)]
capm_top50
```

```python
capm_bottom50=capm_sorted[capm_sorted.Alphas<=capm_sorted.Alphas.quantile(0.1)]
capm_bottom50
```

### Sort capm alpha, buy top 10 and short bottom 10

```python
capm_top10['Tickers'].values
```

```python
capm_bottom10['Tickers'].values
```

```python
port_top10=df_excessrt[capm_top10['Tickers'].values]
port_top10
```

```python
port_bottom10=df_excessrt[capm_bottom10['Tickers'].values]
port_bottom10
```

```python
port_bottom10 = port_bottom10*-1
```

```python
port_first_portfoloio = pd.concat([port_top10,port_bottom10],axis=1)
port_first_portfoloio 
```

```python
avg_returns=port_first_portfoloio.mean()
cov_mat=port_first_portfoloio.cov()
```

```python
np.random.seed(42)
weights = np.random.random(size=(10**5,20))
weights /=  np.sum(weights, axis=1)[:, np.newaxis]
weights
```

```python
###portfolio metrics:
portf_rtns = np.dot(weights, avg_returns)

portf_vol = []
for i in range(0, len(weights)):
    portf_vol.append(np.sqrt(np.dot(weights[i].T, 
                                    np.dot(cov_mat, weights[i]))))
portf_vol = np.array(portf_vol)  
portf_sharpe_ratio = portf_rtns / portf_vol
```

```python
portf_results_df = pd.DataFrame({'returns': portf_rtns,
                                 'volatility': portf_vol,
                                 'sharpe_ratio': portf_sharpe_ratio})
portf_results_df
```

```python
max_sharpe_ind = np.argmax(portf_results_df.sharpe_ratio)
max_sharpe_portf = portf_results_df.loc[max_sharpe_ind]

min_vol_ind = np.argmin(portf_results_df.volatility)
min_vol_portf = portf_results_df.loc[min_vol_ind]
```

```python
RISKY_ASSETS=port_first_portfoloio.columns
RISKY_ASSETS
```

```python
print('Maximum Sharpe Ratio portfolio ----')
print('Performance')
for index, value in max_sharpe_portf.items():
    print(f'{index}: {100 * value:.2f}% ', end="", flush=True)
print('\nWeights')
for x, y in zip(RISKY_ASSETS, weights[np.argmax(portf_results_df.sharpe_ratio)]):
    print(f'{x}: {100*y:.2f}% ', end="", flush=True)
```

```python
print('Minimum Volatility portfolio ----')
print('Performance')
for index, value in min_vol_portf.items():
    print(f'{index}: {100 * value:.2f}% ', end="", flush=True)
print('\nWeights')
for x, y in zip(RISKY_ASSETS, weights[np.argmin(portf_results_df.volatility)]):
    print(f'{x}: {100*y:.2f}% ', end="", flush=True)
```

```python
RISKY_ASSETS=three_factor_first_portfoloio.columns
RISKY_ASSETS
```

```python
print('Maximum Sharpe Ratio portfolio ----')
print('Performance')
for index, value in max_sharpe_portf.items():
    print(f'{index}: {100 * value:.2f}% ', end="", flush=True)
print('\nWeights')
for x, y in zip(RISKY_ASSETS, weights[np.argmax(portf_results_df.sharpe_ratio)]):
    print(f'{x}: {100*y:.2f}% ', end="", flush=True)
```

```python
print('Minimum Volatility portfolio ----')
print('Performance')
for index, value in min_vol_portf.items():
    print(f'{index}: {100 * value:.2f}% ', end="", flush=True)
print('\nWeights')
for x, y in zip(RISKY_ASSETS, weights[np.argmin(portf_results_df.volatility)]):
    print(f'{x}: {100*y:.2f}% ', end="", flush=True)
```

## GQ and White Test

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import datetime

%matplotlib inline

import statsmodels.api as sm
```

```python
df = pd.read_csv('raw_data.csv')
df.head()
```

```python
df = df.set_index(pd.DatetimeIndex(df['Date']))
```

```python
start=datetime.date(1999,4,1)
end=datetime.date(2018,12,1)

rf=web.DataReader('^IRX','yahoo',start=start,end=end)
rf=rf['Adj Close']
rf=(1+(rf/100))**(1/12)-1
rf=rf.resample('M').mean()
rf
```

```python
df=df.resample('M').mean()
df_monthrt=df.pct_change().fillna(0)
df_monthrt
```

```python
df_excessrt= df_monthrt.apply(lambda x: x-rf)
df_excessrt=df_excessrt.apply(lambda x: np.nan_to_num(np.array(x)))
df_excessrt
```

```python
merged_new_factors=pd.read_csv('merged_new_factors.csv')
merged_new_factors

```

```python
merged_new_factors=merged_new_factors.set_index(pd.DatetimeIndex(merged_new_factors['Date']))
merged_new_factors
```

```python
spx=web.DataReader('SPY','yahoo',start=start,end=end)
spx=spx['Adj Close']
spx=spx.resample('M').mean()
spx_rt=spx.pct_change()
spx_rt=spx_rt.fillna(0)
spx_rt
```

```python
result = sm.OLS(spx_rt, sm.add_constant(merged_new_factors[['Pe_ratio','Net_Profit_Margin','ROE',
                                                            'Dividend','Max_Drawdown','^VIX']])).fit()
print(result.summary())
```

```python
from statsmodels.compat import lzip
```

```python
result = sm.OLS(spx_rt, sm.add_constant(merged_new_factors[['Pe_ratio','Net_Profit_Margin','ROE',
                                                            'Dividend','Max_Drawdown','^VIX']]), missing='drop').fit()
result.summary()
```

```python
import statsmodels.stats.api as sms
```

```python
GQ = sms.het_goldfeldquandt(result.resid, result.model.exog)
labels=['Fstat', 'pval']
lzip(labels, GQ)
```

```python
from statsmodels.stats.diagnostic import het_white
```

```python
wtest = het_white(result.resid, result.model.exog)
labels = ['Lagrange Multiplier statistic:', 'LM test\'s p-value:', 'F-statistic:', 'F-test\'s p-value:']
lzip(labels, wtest)
```
